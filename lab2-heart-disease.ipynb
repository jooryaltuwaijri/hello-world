This project uses the Heart Disease Dataset to address a supervised classification problem. The objective is to predict whether a patient has heart disease based on medical attributes such as age, sex, cholesterol level, resting blood pressure, and maximum heart rate. The target variable is target, where a value of 1 indicates the presence of heart disease and 0 indicates its absence.
